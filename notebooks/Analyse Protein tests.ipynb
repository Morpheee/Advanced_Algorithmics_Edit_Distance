```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
from sklearn.metrics import r2_score

from os import listdir
from os.path import isfile, join
```

```python
results_path = 'D:\\Pablo\\clases\\UJM\\1. Semester, 2020\\Advanced Algorithms\\Project\\code\\Advanced_Algorithmics_Edit_Distance\\protein_results'

file_list = [f for f in listdir(results_path) if isfile(join(results_path, f))]

frames = []
for file_name in file_list:
    if file_name.startswith('protein'):
        splitted_name = file_name.split('_')
        df = pd.read_csv(results_path+'/'+file_name)
        df['dataset_length'] = int(splitted_name[1].split('.')[0])
        frames.append(df)
results = pd.concat(frames)

results.head()

#results = results[results['algorithm'] != 'greedy_approach']
```

```python
run_time = results.groupby(['algorithm','dataset_length']).sum().reset_index()[['algorithm','dataset_length','clock_time']]

run_time.pivot(index=['algorithm'], columns=['dataset_length'],values='clock_time').reset_index()
```

```python
run_ed = results.groupby(['algorithm','dataset_length']).sum().reset_index()[['algorithm','dataset_length','ed']]

#run_ed['ed'] = run_ed.apply(lambda row:row['ed']/row['dataset_length'], axis=1)

run_ed.pivot(index=['algorithm'], columns=['dataset_length'],values='ed').reset_index()
```
